# Beautify Soup - Enron Events from web

```python
# import libraries
import urllib.request as urllib2
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# specify the url
quote_page = "https://www.nytimes.com/2006/01/18/business/worldbusiness/timeline-a-chronology-of-enron-corp.html"
```

```python
# query the website and return the html to the variable ‘page’
page = urllib2.urlopen(quote_page)
```

```python
# parse the html using beautiful soup and store in variable `soup`
soup = BeautifulSoup(page, "html.parser")
```

```python
# extract the raw html content from the web
enron_timeline_story_body = soup.find(class_="story-body")
```

```python
# extract the enron events chronology in list
enron_timeline_raw = enron_timeline_story_body.find_all("p")
```

```python
# start from the third element which is the first event for enron corp, get the raw text description
enron_event_desc_content_list = [event_html.get_text() for event_html in enron_timeline_raw[2:]]
```

```python
# identify the four year digit
def isFourYearDigit(str):
    str = str.replace(':', '')

    return len(str) < 5 and str[0:4].isdigit()
```

```python
# method to return the list of time when event happens and another corresponding list of event descriptions
def get_event_timeline(events):
    date_list = []
    desc_list = []
    year = None

    for event in events:

        event_with_date = event.split("-")
        date = None
        desc = None
    
        if len(event_with_date) == 1 and isFourYearDigit(event_with_date[0]):
            year = event_with_date[0]
            continue;
        
        if len(event_with_date) == 1:
            print(event_with_date)
            date = date_list[-1]
            desc = event_with_date[0]
            continue;
        
        if year:
            date = year + ' ' + event_with_date[0]
        else:
            date = event_with_date[0]
        desc = event_with_date[1]
        
        date_list.append(date)
        desc_list.append(desc)

    return date_list, desc_list
```

```python
date_list, desc_list = get_event_timeline(enron_event_desc_content_list)   
```

```python
enron_events = pd.DataFrame({
            "date": date_list,
            "desc": desc_list,
         })   
```
